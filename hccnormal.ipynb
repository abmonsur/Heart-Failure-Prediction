```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix

hcc="../input/hcc-survival-data-set/hcc.csv"
data=pd.read_csv (hcc)
data=data.replace('?',np.NaN)
```

```python
data=data.astype(float)
print(data.dtypes)
```

```python
data.mean()
```

```python
data=data.drop(['5.HepB','9.End','10.Smo','18.EVar','25.Alcpd','26.cigpy','45.dnod','46.Bil','47.Iro','48.Oxy','49.Fer'],axis=1)
data=data.fillna(value=data.mean())
x=data.drop('Class',axis=1)
y=data['Class']
data
```

```python
model=DecisionTreeClassifier()
y_pred = cross_val_predict(model,x, y, cv=10)
met = confusion_matrix(y, y_pred)
print(met)

tp=met[0][0]
fp=met[0][1]
fn=met[1][0]
tn=met[1][1]
pr=tp/(tp+fp)
rec=tp/(tp+fn)
f1=2*pr*rec/(pr+rec)
acc=(tp+tn)/(tp+fp+tn+fn)
print("DecisionTree accuracy:",acc)
print("DecisionTree precision:",pr)
print("DecisionTree recall:",rec)
print("DecisionTree f1 score:",f1)
```

```python
model1=RandomForestClassifier()
y1_pred = cross_val_predict(model1,x, y, cv=10)
met = confusion_matrix(y, y1_pred)
print(met)

tp=met[0][0]
fp=met[0][1]
fn=met[1][0]
tn=met[1][1]
pr=tp/(tp+fp)
rec=tp/(tp+fn)
f1=2*pr*rec/(pr+rec)
acc=(tp+tn)/(tp+fp+tn+fn)
print("RFC accuracy:",acc)
print("RFC precision:",pr)
print("RFC recall:",rec)
print("RFC f1 score:",f1)
```

```python
model2=GaussianNB()
y2_pred = cross_val_predict(model2,x, y, cv=10)
met = confusion_matrix(y, y2_pred)
print(met)

tp=met[0][0]
fp=met[0][1]
fn=met[1][0]
tn=met[1][1]
pr=tp/(tp+fp)
rec=tp/(tp+fn)
f1=2*pr*rec/(pr+rec)
acc=(tp+tn)/(tp+fp+tn+fn)
print("GNB accuracy:",acc)
print("GNB precision:",pr)
print("GNB recall:",rec)
print("GNB f1 score:",f1)
```

```python
model3=LogisticRegression()
y3_pred = cross_val_predict(model3,x, y, cv=10)
met = confusion_matrix(y, y3_pred)
print(met)

tp=met[0][0]
fp=met[0][1]
fn=met[1][0]
tn=met[1][1]
pr=tp/(tp+fp)
rec=tp/(tp+fn)
f1=2*pr*rec/(pr+rec)
acc=(tp+tn)/(tp+fp+tn+fn)
print("LR accuracy:",acc)
print("LR precision:",pr)
print("LR recall:",rec)
print("LR f1 score:",f1)
```

```python
model4=KNeighborsClassifier()
y4_pred = cross_val_predict(model4,x, y, cv=10)
met = confusion_matrix(y, y4_pred)
print(met)

tp=met[0][0]
fp=met[0][1]
fn=met[1][0]
tn=met[1][1]
pr=tp/(tp+fp)
rec=tp/(tp+fn)
f1=2*pr*rec/(pr+rec)
acc=(tp+tn)/(tp+fp+tn+fn)
print("KNN accuracy:",acc)
print("KNN precision:",pr)
print("KNN recall:",rec)
print("KNN f1 score:",f1)
```

```python
model5=SVC()
y5_pred = cross_val_predict(model5,x, y, cv=10)
met = confusion_matrix(y, y5_pred)
print(met)

tp=met[0][0]
fp=met[0][1]
fn=met[1][0]
tn=met[1][1]
pr=tp/(tp+fp)
rec=tp/(tp+fn)
f1=2*pr*rec/(pr+rec)
acc=(tp+tn)/(tp+fp+tn+fn)
print("SVC accuracy:",acc)
print("SVC precision:",pr)
print("SVC recall:",rec)
print("SVC f1 score:",f1)
```
